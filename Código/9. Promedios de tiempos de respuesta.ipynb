```python
# √çtems de cambio de Tiempo progresistas.
Items_Progresistas_Izquierda_Tiempo = ['IP_Item_5_Izq_Tiempo', 
                                       'IP_Item_6_Izq_Tiempo', 
                                       'IP_Item_9_Izq_Tiempo', 
                                       'IP_Item_11_Izq_Tiempo', 
                                       'IP_Item_16_Izq_Tiempo', 
                                       'IP_Item_20_Izq_Tiempo', 
                                       'IP_Item_24_Izq_Tiempo', 
                                       'IP_Item_25_Izq_Tiempo', 
                                       'IP_Item_27_Izq_Tiempo', 
                                       'IP_Item_28_Izq_Tiempo']

Items_Progresistas_Derecha_Tiempo = ['IP_Item_5_Der_Tiempo', 
                                     'IP_Item_6_Der_Tiempo', 
                                     'IP_Item_9_Der_Tiempo', 
                                     'IP_Item_11_Der_Tiempo', 
                                     'IP_Item_16_Der_Tiempo', 
                                     'IP_Item_20_Der_Tiempo', 
                                     'IP_Item_24_Der_Tiempo', 
                                     'IP_Item_25_Der_Tiempo', 
                                     'IP_Item_27_Der_Tiempo', 
                                     'IP_Item_28_Der_Tiempo']
```

```python
# √çtems de cambio de Tiempo conservadores.
Items_Conservadores_Izquierda_Tiempo = ['IP_Item_3_Izq_Tiempo', 
                                        'IP_Item_4_Izq_Tiempo', 
                                        'IP_Item_7_Izq_Tiempo', 
                                        'IP_Item_8_Izq_Tiempo', 
                                        'IP_Item_10_Izq_Tiempo', 
                                        'IP_Item_19_Izq_Tiempo', 
                                        'IP_Item_22_Izq_Tiempo', 
                                        'IP_Item_23_Izq_Tiempo', 
                                        'IP_Item_29_Izq_Tiempo', 
                                        'IP_Item_30_Izq_Tiempo']

Items_Conservadores_Derecha_Tiempo = ['IP_Item_3_Der_Tiempo', 
                                     'IP_Item_4_Der_Tiempo', 
                                     'IP_Item_7_Der_Tiempo', 
                                     'IP_Item_8_Der_Tiempo', 
                                     'IP_Item_10_Der_Tiempo', 
                                     'IP_Item_19_Der_Tiempo', 
                                     'IP_Item_22_Der_Tiempo', 
                                     'IP_Item_23_Der_Tiempo', 
                                     'IP_Item_29_Der_Tiempo', 
                                     'IP_Item_30_Der_Tiempo']
```

```python
for Nombre_df, df in dfs_Finales.items():
   df['Tiempos_Respuesta_Promedio_Pro_Izq'] = df[Items_Progresistas_Izquierda_Tiempo].mean(axis=1)
   df['Tiempos_Respuesta_Promedio_Pro_Der'] = df[Items_Progresistas_Derecha_Tiempo].mean(axis=1)
   df['Tiempos_Respuesta_Promedio_Con_Izq'] = df[Items_Conservadores_Izquierda_Tiempo].mean(axis=1)
   df['Tiempos_Respuesta_Promedio_Con_Der'] = df[Items_Conservadores_Derecha_Tiempo].mean(axis=1)

   df['Tiempos_Respuesta_Promedio_Pro'] = (df['Tiempos_Respuesta_Promedio_Pro_Izq'] + df['Tiempos_Respuesta_Promedio_Pro_Der']) / 2
   df['Tiempos_Respuesta_Promedio_Con'] = (df['Tiempos_Respuesta_Promedio_Con_Izq'] + df['Tiempos_Respuesta_Promedio_Con_Der']) / 2
```

```python
print("="*80)
print("VERIFICACI√ìN: PROMEDIOS DE TIEMPOS DE RESPUESTA")
print("="*80)

# 1. Verificar existencia de listas de √≠tems.
print("1. VERIFICACI√ìN DE LISTAS DE √çTEMS:")

Listas_Requeridas = {
    'Items_Progresistas_Izquierda_Tiempo': Items_Progresistas_Izquierda_Tiempo,
    'Items_Progresistas_Derecha_Tiempo': Items_Progresistas_Derecha_Tiempo,
    'Items_Conservadores_Izquierda_Tiempo': Items_Conservadores_Izquierda_Tiempo,
    'Items_Conservadores_Derecha_Tiempo': Items_Conservadores_Derecha_Tiempo
}

for Nombre_Lista, Lista in Listas_Requeridas.items():
    print(f"   ‚úÖ {Nombre_Lista}: {len(Lista)} √≠tems")
    print(f"      Primeros 3: {Lista[:3]}")

print("\n" + "="*60)

for Nombre_df, df in dfs_Finales.items():
    print(f"\nüìä DATAFRAME: {Nombre_df}")
    print("-" * 50)
    
    # 2. Verificar existencia de columnas de tiempo.
    print("2. VERIFICACI√ìN DE COLUMNAS DE TIEMPO:")
    
    Todas_Columnas_Tiempo = (Items_Progresistas_Izquierda_Tiempo + 
                            Items_Progresistas_Derecha_Tiempo +
                            Items_Conservadores_Izquierda_Tiempo + 
                            Items_Conservadores_Derecha_Tiempo)
    
    Columnas_Encontradas = []
    Columnas_Faltantes = []
    
    for Columna in Todas_Columnas_Tiempo:
        if Columna in df.columns:
            Columnas_Encontradas.append(Columna)
        else:
            Columnas_Faltantes.append(Columna)
    
    print(f"   ‚úÖ Columnas encontradas: {len(Columnas_Encontradas)}/{len(Todas_Columnas_Tiempo)}")
    print(f"   ‚ùå Columnas faltantes: {len(Columnas_Faltantes)}")
    
    if len(Columnas_Faltantes) > 0:
        print(f"      Faltantes: {Columnas_Faltantes[:5]}...")  # Mostrar primeras 5.
    
    # 3. Verificar creaci√≥n de columnas de promedios.
    print("\n3. VERIFICACI√ìN DE COLUMNAS DE PROMEDIOS CREADAS:")
    
    Columnas_Promedios_Esperadas = [
        'Tiempos_Respuesta_Promedio_Pro_Izq',
        'Tiempos_Respuesta_Promedio_Pro_Der', 
        'Tiempos_Respuesta_Promedio_Con_Izq',
        'Tiempos_Respuesta_Promedio_Con_Der',
        'Tiempos_Respuesta_Promedio_Pro',
        'Tiempos_Respuesta_Promedio_Con'
    ]
    
    for Columna in Columnas_Promedios_Esperadas:
        if Columna in df.columns:
            Datos_Validos = df[Columna].notna().sum()
            Total_Casos = len(df)
            Porcentaje = (Datos_Validos / Total_Casos) * 100
            print(f"   ‚úÖ {Columna}: {Datos_Validos}/{Total_Casos} ({Porcentaje:.1f}%)")
        else:
            print(f"   ‚ùå {Columna}: NO CREADA")
    
    # 4. Estad√≠sticas descriptivas de los promedios.
    print("\n4. ESTAD√çSTICAS DESCRIPTIVAS:")
    
    for Columna in Columnas_Promedios_Esperadas:
        if Columna in df.columns:
            Stats = df[Columna].describe()
            print(f"\n   üìä {Columna}:")
            print(f"      Count: {Stats['count']:.0f}")
            print(f"      Mean: {Stats['mean']:.3f}")
            print(f"      Std: {Stats['std']:.3f}")
            print(f"      Min: {Stats['min']:.3f}")
            print(f"      Max: {Stats['max']:.3f}")
            
            # Verificar valores an√≥malos.
            if Stats['min'] < 0:
                print(f"      ‚ö†Ô∏è  Tiempos negativos detectados")
            
            if Stats['max'] > 300:  # M√°s de 5 minutos parece excesivo.
                print(f"      ‚ö†Ô∏è  Tiempos muy largos detectados (>{Stats['max']:.1f}s)")
    
    # 5. Verificaci√≥n manual de c√°lculos.
    print("\n5. VERIFICACI√ìN MANUAL DE C√ÅLCULOS:")
    
    # Tomar una muestra para verificar c√°lculos.
    if len(df) > 0:
        Muestra = df.dropna(subset=['Tiempos_Respuesta_Promedio_Pro']).iloc[0] if 'Tiempos_Respuesta_Promedio_Pro' in df.columns else df.iloc[0]
        ID_Muestra = Muestra['ID']
        
        print(f"   Verificando participante ID: {ID_Muestra}")
        
        # Verificar c√°lculo de promedio progresista izquierda.
        if 'Tiempos_Respuesta_Promedio_Pro_Izq' in df.columns:
            Promedio_Pro_Izq_Calculado = Muestra['Tiempos_Respuesta_Promedio_Pro_Izq']
            
            # Calcular manualmente.
            Valores_Pro_Izq = []
            for Item in Items_Progresistas_Izquierda_Tiempo:
                if Item in df.columns:
                    Valor = Muestra[Item]
                    if pd.notna(Valor):
                        Valores_Pro_Izq.append(Valor)
            
            if len(Valores_Pro_Izq) > 0:
                Promedio_Manual = sum(Valores_Pro_Izq) / len(Valores_Pro_Izq)
                print(f"   Pro_Izq calculado: {Promedio_Pro_Izq_Calculado:.3f}")
                print(f"   Pro_Izq manual: {Promedio_Manual:.3f}")
                
                Diferencia = abs(Promedio_Pro_Izq_Calculado - Promedio_Manual) if pd.notna(Promedio_Pro_Izq_Calculado) else float('inf')
                if Diferencia < 0.001:
                    print("   ‚úÖ C√°lculo Pro_Izq correcto")
                else:
                    print(f"   ‚ö†Ô∏è  Diferencia Pro_Izq: {Diferencia:.3f}")
        
        # Verificar c√°lculo de promedio final progresista.
        if ('Tiempos_Respuesta_Promedio_Pro' in df.columns and
            'Tiempos_Respuesta_Promedio_Pro_Izq' in df.columns and
            'Tiempos_Respuesta_Promedio_Pro_Der' in df.columns):
            
            Promedio_Final_Calculado = Muestra['Tiempos_Respuesta_Promedio_Pro']
            Pro_Izq = Muestra['Tiempos_Respuesta_Promedio_Pro_Izq']
            Pro_Der = Muestra['Tiempos_Respuesta_Promedio_Pro_Der']
            
            if pd.notna(Pro_Izq) and pd.notna(Pro_Der):
                Promedio_Final_Manual = (Pro_Izq + Pro_Der) / 2
                print(f"   Pro_Final calculado: {Promedio_Final_Calculado:.3f}")
                print(f"   Pro_Final manual: {Promedio_Final_Manual:.3f}")
                
                Diferencia = abs(Promedio_Final_Calculado - Promedio_Final_Manual) if pd.notna(Promedio_Final_Calculado) else float('inf')
                if Diferencia < 0.001:
                    print("   ‚úÖ C√°lculo Pro_Final correcto")
                else:
                    print(f"   ‚ö†Ô∏è  Diferencia Pro_Final: {Diferencia:.3f}")
    
    # 6. Distribuci√≥n por categor√≠as.
    print("\n6. DISTRIBUCI√ìN POR CATEGOR√çAS PASO:")
    
    if 'Categoria_PASO_2023' in df.columns:
        Columnas_Para_Analizar = ['Tiempos_Respuesta_Promedio_Pro', 'Tiempos_Respuesta_Promedio_Con']
        
        for Columna in Columnas_Para_Analizar:
            if Columna in df.columns:
                print(f"\n   üìä {Columna}:")
                print("   Categor√≠a               | Count | Mean    | Std     |")
                print("   " + "-"*55)
                
                Distribucion = df.groupby('Categoria_PASO_2023')[Columna].agg(['count', 'mean', 'std'])
                
                for Categoria, Stats in Distribucion.iterrows():
                    if pd.notna(Stats['mean']) and Stats['count'] > 0:
                        print(f"   {Categoria:<22} | {Stats['count']:5.0f} | {Stats['mean']:7.2f} | {Stats['std']:7.2f} |")
    
    # 7. An√°lisis de completitud de datos.
    print("\n7. AN√ÅLISIS DE COMPLETITUD:")
    
    # Verificar qu√© tanto influyen los datos faltantes.
    Estadisticas_Completitud = {}
    
    for Grupo, Lista_Items in Listas_Requeridas.items():
        Columnas_Existentes = [col for col in Lista_Items if col in df.columns]
        
        if len(Columnas_Existentes) > 0:
            # Contar casos con al menos un dato v√°lido.
            Al_Menos_Uno = (df[Columnas_Existentes].notna().sum(axis=1) > 0).sum()
            # Contar casos con todos los datos v√°lidos.
            Todos_Validos = (df[Columnas_Existentes].notna().sum(axis=1) == len(Columnas_Existentes)).sum()
            
            Estadisticas_Completitud[Grupo] = {
                'Columnas_Disponibles': len(Columnas_Existentes),
                'Al_Menos_Uno': Al_Menos_Uno,
                'Todos_Validos': Todos_Validos,
                'Total_Casos': len(df)
            }
    
    print("   Grupo                           | Cols | ‚â•1 Dato | Completos | Total |")
    print("   " + "-"*75)
    
    for Grupo, Stats in Estadisticas_Completitud.items():
        Grupo_Corto = Grupo.replace('Items_', '').replace('_Tiempo', '')
        print(f"   {Grupo_Corto:<30} | {Stats['Columnas_Disponibles']:4} | {Stats['Al_Menos_Uno']:7} | {Stats['Todos_Validos']:9} | {Stats['Total_Casos']:5} |")
    
    # 8. Detecci√≥n de outliers en tiempos.
    print("\n8. DETECCI√ìN DE OUTLIERS:")
    
    for Columna in ['Tiempos_Respuesta_Promedio_Pro', 'Tiempos_Respuesta_Promedio_Con']:
        if Columna in df.columns:
            Q1 = df[Columna].quantile(0.25)
            Q3 = df[Columna].quantile(0.75)
            IQR = Q3 - Q1
            Limite_Inferior = Q1 - 1.5 * IQR
            Limite_Superior = Q3 + 1.5 * IQR
            
            Outliers_Inferiores = (df[Columna] < Limite_Inferior).sum()
            Outliers_Superiores = (df[Columna] > Limite_Superior).sum()
            
            print(f"   {Columna}:")
            print(f"     L√≠mites IQR: [{Limite_Inferior:.2f}, {Limite_Superior:.2f}]")
            print(f"     Outliers inferiores: {Outliers_Inferiores}")
            print(f"     Outliers superiores: {Outliers_Superiores}")
            
            if Outliers_Superiores > 0:
                Valor_Max_Outlier = df[df[Columna] > Limite_Superior][Columna].max()
                print(f"     Valor m√°ximo outlier: {Valor_Max_Outlier:.2f}s")

print("\n" + "="*80)
print("‚úÖ VERIFICACI√ìN DE PROMEDIOS DE TIEMPOS COMPLETADA")
print("="*80)

# 9. Verificaci√≥n adicional: Coherencia entre progresistas y conservadores.
print("\n" + "="*80)
print("VERIFICACI√ìN ADICIONAL: COHERENCIA TEMPORAL")
print("="*80)

for Nombre_df, df in dfs_Finales.items():
    print(f"\nüïí DATAFRAME: {Nombre_df}")
    print("-" * 40)
    
    if ('Tiempos_Respuesta_Promedio_Pro' in df.columns and
        'Tiempos_Respuesta_Promedio_Con' in df.columns):
        
        # Correlaci√≥n entre tiempos progresistas y conservadores.
        Correlacion_Tiempos = df['Tiempos_Respuesta_Promedio_Pro'].corr(df['Tiempos_Respuesta_Promedio_Con'])
        print(f"   Correlaci√≥n Pro vs Con: {Correlacion_Tiempos:.3f}")
        
        if Correlacion_Tiempos > 0.5:
            print("   ‚úÖ Correlaci√≥n positiva fuerte (esperado)")
        elif Correlacion_Tiempos > 0.2:
            print("   ‚ö†Ô∏è  Correlaci√≥n positiva moderada")
        else:
            print("   ‚ùå Correlaci√≥n d√©bil o negativa")
        
        # Comparar tiempos promedio generales.
        Tiempo_Pro_Medio = df['Tiempos_Respuesta_Promedio_Pro'].mean()
        Tiempo_Con_Medio = df['Tiempos_Respuesta_Promedio_Con'].mean()
        
        print(f"   Tiempo promedio Progresistas: {Tiempo_Pro_Medio:.2f}s")
        print(f"   Tiempo promedio Conservadores: {Tiempo_Con_Medio:.2f}s")
        
        Diferencia_Porcentual = abs(Tiempo_Pro_Medio - Tiempo_Con_Medio) / ((Tiempo_Pro_Medio + Tiempo_Con_Medio) / 2) * 100
        print(f"   Diferencia porcentual: {Diferencia_Porcentual:.1f}%")
        
        if Diferencia_Porcentual < 10:
            print("   ‚úÖ Tiempos similares entre tipos de √≠tems")
        else:
            print("   ‚ö†Ô∏è  Diferencia significativa en tiempos")

print("\n" + "="*80)
print("‚úÖ VERIFICACI√ìN DE COHERENCIA TEMPORAL COMPLETADA")
print("="*80)
```
