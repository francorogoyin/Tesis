```python
print("="*80)
print("VERIFICACIÓN: AGRUPAMIENTOS DE VARIABLES SOCIOECONÓMICAS")
print("="*80)

# 1. Verificar existencia de diccionarios de mapeo.
print("1. VERIFICACIÓN DE DICCIONARIOS DE MAPEO:")

Diccionarios_Mapeo = [
    ('Region', 'Region'),
    ('AID', 'Autopercepción Izq-Der'),
    ('ACP', 'Autopercepción Con-Pro'), 
    ('APA', 'Autopercepción Per-Antiper'),
    ('Mapeo_Genero', 'Género'),
    ('Mapeo_Inmueble', 'Inmueble Residencia'),
    ('Mapeo_Fuente_Ingreso', 'Fuente de Ingreso'),
    ('Mapeo_Estrato_Social', 'Estrato Social'),
    ('Mapeo_Nivel_Educativo', 'Nivel Educativo'),
    ('Mapeo_Votara_2023', 'Votará 2023'),
    ('Mapeo_Voto_2019', 'Voto 2019'),
    ('Mapeo_Voto_PASO_2023', 'Voto PASO 2023'),
    ('Mapeo_Candidato_PASO_2023', 'Candidato PASO 2023'),
    ('Mapeo_Categoria_PASO_2023', 'Categoría PASO 2023'),
    ('Mapeo_Afiliacion_Politica', 'Afiliación Política')
]

for Variable_Dict, Nombre_Descriptivo in Diccionarios_Mapeo:
    if Variable_Dict in globals():
        Diccionario = globals()[Variable_Dict]
        print(f"   ✅ {Nombre_Descriptivo}: {len(Diccionario)} mapeos definidos")
    else:
        print(f"   ❌ {Nombre_Descriptivo}: Diccionario no encontrado")

print("\n" + "="*60)

for Nombre_df, df in dfs_Finales.items():
    print(f"\n🏛️ DATAFRAME: {Nombre_df}")
    print("-" * 50)
    
    Total_Casos = len(df)
    
    # 2. Verificar mapeo de regiones.
    print("2. VERIFICACIÓN DE REGIONES:")
    
    if 'Region' in df.columns:
        print("   ✅ Columna 'Region' creada")
        
        Distribucion_Regiones = df['Region'].value_counts()
        print("   Distribución por región:")
        
        for Region, Cantidad in Distribucion_Regiones.items():
            Porcentaje = (Cantidad / Total_Casos) * 100
            print(f"     {Region:<15}: {Cantidad:4} ({Porcentaje:5.1f}%)")
        
        # Verificar casos sin región.
        Sin_Region = (df['Region'] == 'Sin region').sum()
        if Sin_Region > 0:
            print(f"   ⚠️  {Sin_Region} casos 'Sin region' - revisar provincias no mapeadas")
            if 'Provincia' in df.columns:
                Provincias_Sin_Mapear = df[df['Region'] == 'Sin region']['Provincia'].value_counts()
                print("     Provincias no mapeadas:")
                for Prov, Cant in Provincias_Sin_Mapear.items():
                    print(f"       '{Prov}': {Cant} casos")
        else:
            print("   ✅ Todas las provincias mapeadas correctamente")
    else:
        print("   ❌ Columna 'Region' no encontrada")
    
    # 3. Verificar agrupaciones de autopercepción.
    print("\n3. VERIFICACIÓN DE AUTOPERCEPCIÓN AGRUPADA:")
    
    Variables_Autopercepcion = [
        ('Autopercepcion_Izq_Der_Agrupada', 'Autopercepcion_Izq_Der'),
        ('Autopercepcion_Con_Pro_Agrupada', 'Autopercepcion_Con_Pro'),
        ('Autopercepcion_Per_Antiper_Agrupada', 'Autopercepcion_Per_Antiper')
    ]
    
    for Var_Agrupada, Var_Original in Variables_Autopercepcion:
        if Var_Agrupada in df.columns:
            print(f"\n   📊 {Var_Agrupada}:")
            
            Distribucion = df[Var_Agrupada].value_counts()
            for Categoria, Cantidad in Distribucion.items():
                Porcentaje = (Cantidad / Total_Casos) * 100
                print(f"     {Categoria}: {Cantidad:4} ({Porcentaje:5.1f}%)")
            
            # Verificar casos "Otro".
            Casos_Otro = (df[Var_Agrupada] == 'Otro').sum()
            if Casos_Otro > 0:
                print(f"   ⚠️  {Casos_Otro} casos 'Otro' - verificar valores originales")
                if Var_Original in df.columns:
                    Valores_Otro = df[df[Var_Agrupada] == 'Otro'][Var_Original].value_counts()
                    print("     Valores originales no mapeados:")
                    for Valor, Cant in Valores_Otro.head(5).items():
                        print(f"       {Valor}: {Cant} casos")
        else:
            print(f"   ❌ {Var_Agrupada} no encontrada")
    
    # 4. Verificar agrupación de edades.
    print("\n4. VERIFICACIÓN DE EDADES AGRUPADAS:")
    
    if 'Edad_Agrupada' in df.columns:
        print("   ✅ Columna 'Edad_Agrupada' creada")
        
        Distribucion_Edades = df['Edad_Agrupada'].value_counts().sort_index()
        print("   Distribución por rango de edad:")
        
        for Rango, Cantidad in Distribucion_Edades.items():
            if pd.notna(Rango):
                Porcentaje = (Cantidad / Total_Casos) * 100
                print(f"     {Rango}: {Cantidad:4} ({Porcentaje:5.1f}%)")
        
        # Verificar casos sin edad.
        Sin_Edad = df['Edad_Agrupada'].isna().sum()
        if Sin_Edad > 0:
            print(f"   ⚠️  {Sin_Edad} casos sin edad válida")
        
        # Verificar coherencia con edad original.
        if 'Edad' in df.columns:
            print("   Verificación de rangos:")
            for Rango in ['19_35', '36_50', '51_70']:
                if Rango in df['Edad_Agrupada'].values:
                    Muestra = df[df['Edad_Agrupada'] == Rango]['Edad'].dropna()
                    if len(Muestra) > 0:
                        Min_Edad = Muestra.min()
                        Max_Edad = Muestra.max()
                        print(f"     {Rango}: Edades {Min_Edad}-{Max_Edad}")
    else:
        print("   ❌ Columna 'Edad_Agrupada' no encontrada")
    
    # 5. VERIFICACIÓN MANUAL DE CASOS ESPECÍFICOS.
    print("\n5. VERIFICACIÓN MANUAL - CASOS ESPECÍFICOS:")
    
    # Seleccionar 5 casos al azar para verificación manual.
    if len(df) >= 5:
        import random
        random.seed(456)  # Para reproducibilidad.
        Indices_Muestra = random.sample(range(len(df)), min(5, len(df)))
        
        Variables_A_Verificar = ['Genero', 'Inmueble_Residencia', 'Fuente_Ingreso', 
                               'Nivel_Educativo', 'Estrato_Social']
        
        for i, Indice in enumerate(Indices_Muestra, 1):
            Fila = df.iloc[Indice]
            ID_Participante = Fila['ID']
            
            print(f"\n   🔍 CASO {i} - ID: {ID_Participante}")
            
            for Variable in Variables_A_Verificar:
                if Variable in df.columns:
                    Valor_Final = Fila[Variable]
                    print(f"      {Variable}: '{Valor_Final}'")
                    
                    # Verificar si quedó como "Otro".
                    if Valor_Final == 'Otro':
                        print(f"        ⚠️  Marcado como 'Otro' - revisar valor original")
            
            # Verificar autopercepción.
            if 'Autopercepcion_Izq_Der_Agrupada' in df.columns:
                Valor_Agrupado = Fila['Autopercepcion_Izq_Der_Agrupada']
                Valor_Original = Fila.get('Autopercepcion_Izq_Der', 'N/A')
                print(f"      Autopercepción Izq-Der: {Valor_Original} → {Valor_Agrupado}")
            
            # Verificar región.
            if 'Region' in df.columns and 'Provincia' in df.columns:
                Provincia = Fila['Provincia']
                Region = Fila['Region']
                print(f"      Región: '{Provincia}' → '{Region}'")
            
            # Verificar edad agrupada.
            if 'Edad_Agrupada' in df.columns and 'Edad' in df.columns:
                Edad = Fila['Edad']
                Rango = Fila['Edad_Agrupada']
                print(f"      Edad: {Edad} → {Rango}")
    
    # 6. Análisis de completitud de variables.
    print("\n6. ANÁLISIS DE COMPLETITUD:")
    
    Variables_Criticas = [
        'Region', 'Edad_Agrupada', 'Genero', 'Nivel_Educativo', 
        'Estrato_Social', 'Inmueble_Residencia', 'Fuente_Ingreso'
    ]
    
    print("   Variable                 | Válidos | % Completo | % 'Otro'")
    print("   " + "-"*65)
    
    for Variable in Variables_Criticas:
        if Variable in df.columns:
            Valores_Validos = df[Variable].notna().sum()
            Porcentaje_Completo = (Valores_Validos / Total_Casos) * 100
            
            # Contar casos "Otro".
            Casos_Otro = (df[Variable] == 'Otro').sum() if df[Variable].dtype == 'object' else 0
            Porcentaje_Otro = (Casos_Otro / Total_Casos) * 100
            
            print(f"   {Variable:<23} | {Valores_Validos:7} | {Porcentaje_Completo:9.1f}% | {Porcentaje_Otro:7.1f}%")
        else:
            print(f"   {Variable:<23} | {'N/A':7} | {'N/A':9} | {'N/A':7}")
    
    # 7. Verificar coherencia entre variables relacionadas.
    print("\n7. VERIFICACIÓN DE COHERENCIA:")
    
    # Coherencia entre voto y categoría.
    if ('Candidato_PASO_2023' in df.columns and 'Categoria_PASO_2023' in df.columns):
        print("\n   Coherencia Candidato vs Categoría PASO 2023:")
        
        # Verificar algunos casos específicos.
        Coherencias_Esperadas = {
            'Javier_Milei': 'Right_Wing_Libertarian',
            'Sergio_Massa': 'Progressivism',
            'Patricia_Bullrich': 'Moderate_Right_B',
            'Myriam_Bregman': 'Left_Wing',
            'Horacio_Rodriguez_Larreta': 'Moderate_Right_A',
            'Juan_Grabois': 'Progressivism',
            'Gabriel_Solano': 'Left_Wing',
            'Juan_Schiaretti': 'Centre',
            'Manuela_Castañeira': 'Left_Wing'
        }
        
        for Candidato, Categoria_Esperada in Coherencias_Esperadas.items():
            if Candidato in df['Candidato_PASO_2023'].values:
                Casos_Candidato = df[df['Candidato_PASO_2023'] == Candidato]
                Categorias_Encontradas = Casos_Candidato['Categoria_PASO_2023'].value_counts()
                
                Casos_Coherentes = Casos_Candidato[Casos_Candidato['Categoria_PASO_2023'] == Categoria_Esperada].shape[0]
                Total_Candidato = len(Casos_Candidato)
                
                if Total_Candidato > 0:
                    Porcentaje_Coherencia = (Casos_Coherentes / Total_Candidato) * 100
                    print(f"     {Candidato}: {Porcentaje_Coherencia:.1f}% coherencia ({Casos_Coherentes}/{Total_Candidato})")
    
    # Coherencia nivel educativo vs estrato social.
    if ('Nivel_Educativo' in df.columns and 'Estrato_Social' in df.columns):
        print("\n   Análisis Nivel Educativo vs Estrato Social:")
        
        # Crear tabla cruzada.
        Tabla_Cruzada = pd.crosstab(df['Nivel_Educativo'], df['Estrato_Social'], margins=True)
        
        # Mostrar algunas relaciones clave.
        if 'Posgrado' in df['Nivel_Educativo'].values and 'Alto' in df['Estrato_Social'].values:
            Posgrado_Alto = df[(df['Nivel_Educativo'] == 'Posgrado') & (df['Estrato_Social'] == 'Alto')].shape[0]
            Total_Posgrado = (df['Nivel_Educativo'] == 'Posgrado').sum()
            if Total_Posgrado > 0:
                Porcentaje = (Posgrado_Alto / Total_Posgrado) * 100
                print(f"     Posgrado → Estrato Alto: {Porcentaje:.1f}% ({Posgrado_Alto}/{Total_Posgrado})")
    
    # 8. Detección de inconsistencias.
    print("\n8. DETECCIÓN DE INCONSISTENCIAS:")
    
    Inconsistencias_Encontradas = 0
    
    # Edad vs nivel educativo.
    if ('Edad' in df.columns and 'Nivel_Educativo' in df.columns):
        # Buscar casos de posgrado muy jóvenes.
        Posgrado_Jovenes = df[(df['Nivel_Educativo'] == 'Posgrado') & (df['Edad'] < 25)].shape[0]
        if Posgrado_Jovenes > 0:
            print(f"   ⚠️  {Posgrado_Jovenes} casos de Posgrado menores de 25 años")
            Inconsistencias_Encontradas += Posgrado_Jovenes
    
    # Jubilación vs edad.
    if ('Fuente_Ingreso' in df.columns and 'Edad' in df.columns):
        Jubilados_Jovenes = df[(df['Fuente_Ingreso'] == 'Jubilacion') & (df['Edad'] < 60)].shape[0]
        if Jubilados_Jovenes > 0:
            print(f"   ⚠️  {Jubilados_Jovenes} casos de Jubilación menores de 60 años")
            Inconsistencias_Encontradas += Jubilados_Jovenes
    
    # Estrato alto vs fuente de ingreso.
    if ('Estrato_Social' in df.columns and 'Fuente_Ingreso' in df.columns):
        Alto_Desocupado = df[(df['Estrato_Social'] == 'Alto') & (df['Fuente_Ingreso'] == 'Desocupado')].shape[0]
        if Alto_Desocupado > 0:
            print(f"   ⚠️  {Alto_Desocupado} casos de Estrato Alto + Desocupado")
            Inconsistencias_Encontradas += Alto_Desocupado
    
    if Inconsistencias_Encontradas == 0:
        print("   ✅ No se detectaron inconsistencias obvias")
    
    # 9. Análisis por categoría política.
    print("\n9. PERFIL SOCIOECONÓMICO POR CATEGORÍA POLÍTICA:")
    
    if 'Categoria_PASO_2023' in df.columns:
        Categorias_Principales = ['Left_Wing', 'Progressivism', 'Centre', 
                                'Moderate_Right_A', 'Right_Wing_Libertarian']
        
        Variables_Perfil = ['Nivel_Educativo', 'Estrato_Social', 'Region']
        
        for Variable in Variables_Perfil:
            if Variable in df.columns:
                print(f"\n   {Variable} por categoría política:")
                
                for Categoria in Categorias_Principales:
                    if Categoria in df['Categoria_PASO_2023'].values:
                        df_Cat = df[df['Categoria_PASO_2023'] == Categoria]
                        if len(df_Cat) > 0 and Variable in df_Cat.columns:
                            Valor_Mas_Frecuente = df_Cat[Variable].mode()
                            if len(Valor_Mas_Frecuente) > 0:
                                Frecuencia = (df_Cat[Variable] == Valor_Mas_Frecuente.iloc[0]).sum()
                                Porcentaje = (Frecuencia / len(df_Cat)) * 100
                                print(f"     {Categoria:<20}: {Valor_Mas_Frecuente.iloc[0]} ({Porcentaje:.1f}%)")

print("\n" + "="*80)
print("✅ VERIFICACIÓN DE AGRUPAMIENTOS COMPLETADA")
print("="*80)

# 10. Resumen ejecutivo.
print("\n" + "="*80)
print("RESUMEN EJECUTIVO - VARIABLES SOCIOECONÓMICAS")
print("="*80)

for Nombre_df, df in dfs_Finales.items():
    print(f"\n📊 DATAFRAME: {Nombre_df}")
    
    # Región más representada.
    if 'Region' in df.columns:
        Region_Top = df['Region'].mode().iloc[0] if len(df['Region'].mode()) > 0 else 'N/A'
        Region_Casos = (df['Region'] == Region_Top).sum()
        Region_Porcentaje = (Region_Casos / len(df)) * 100
        print(f"   🗺️  Región principal: {Region_Top} ({Region_Porcentaje:.1f}%)")
    
    # Nivel educativo predominante.
    if 'Nivel_Educativo' in df.columns:
        Educacion_Top = df['Nivel_Educativo'].mode().iloc[0] if len(df['Nivel_Educativo'].mode()) > 0 else 'N/A'
        Educacion_Porcentaje = (df['Nivel_Educativo'] == Educacion_Top).sum() / len(df) * 100
        print(f"   🎓 Nivel educativo principal: {Educacion_Top} ({Educacion_Porcentaje:.1f}%)")
    
    # Estrato social predominante.
    if 'Estrato_Social' in df.columns:
        Estrato_Top = df['Estrato_Social'].mode().iloc[0] if len(df['Estrato_Social'].mode()) > 0 else 'N/A'
        Estrato_Porcentaje = (df['Estrato_Social'] == Estrato_Top).sum() / len(df) * 100
        print(f"   💰 Estrato social principal: {Estrato_Top} ({Estrato_Porcentaje:.1f}%)")
    
    # Rango de edad predominante.
    if 'Edad_Agrupada' in df.columns:
        Edad_Top = df['Edad_Agrupada'].mode().iloc[0] if len(df['Edad_Agrupada'].mode()) > 0 else 'N/A'
        Edad_Porcentaje = (df['Edad_Agrupada'] == Edad_Top).sum() / len(df) * 100
        print(f"   👥 Rango de edad principal: {Edad_Top} ({Edad_Porcentaje:.1f}%)")
    
    # Total de variables procesadas.
    Variables_Procesadas = sum(1 for var in Variables_Criticas if var in df.columns)
    print(f"   🔧 Variables socioeconómicas procesadas: {Variables_Procesadas}")

print("\n" + "="*80)
print("✅ ANÁLISIS SOCIOECONÓMICO COMPLETADO")
print("="*80)
```
