```python
print("="*80)
print("VERIFICACI√ìN: √çNDICE DE CONSERVADURISMO")
print("="*80)

# 1. Verificar existencia de variables necesarias.
print("1. VERIFICACI√ìN DE VARIABLES NECESARIAS:")
Variables_Requeridas = ['Items_Conservadores', 'Items_Conservadores_Tiempo']

for Variable in Variables_Requeridas:
    if Variable in globals():
        Valor_Variable = globals()[Variable]
        print(f"   ‚úÖ {Variable}: {len(Valor_Variable) if hasattr(Valor_Variable, '__len__') else 'Definida'}")
        if hasattr(Valor_Variable, '__len__') and len(Valor_Variable) > 0:
            print(f"      Contenido: {list(Valor_Variable)[:10]}...")  # Primeros 10 elementos.
    else:
        print(f"   ‚ùå {Variable}: NO ENCONTRADA")
        # Definir manualmente si no existe (√≠tems conservadores t√≠picos).
        if Variable == 'Items_Conservadores':
            Items_Conservadores = [
                'IP_Item_3_Respuesta', 'IP_Item_4_Respuesta', 'IP_Item_7_Respuesta',
                'IP_Item_8_Respuesta', 'IP_Item_10_Respuesta', 'IP_Item_19_Respuesta',
                'IP_Item_22_Respuesta', 'IP_Item_23_Respuesta', 'IP_Item_29_Respuesta',
                'IP_Item_30_Respuesta'
            ]
            print(f"      üîß Definiendo manualmente: {Items_Conservadores}")
        elif Variable == 'Items_Conservadores_Tiempo':
            Items_Conservadores_Tiempo = [
                'IP_Item_3_Tiempo', 'IP_Item_4_Tiempo', 'IP_Item_7_Tiempo',
                'IP_Item_8_Tiempo', 'IP_Item_10_Tiempo', 'IP_Item_19_Tiempo',
                'IP_Item_22_Tiempo', 'IP_Item_23_Tiempo', 'IP_Item_29_Tiempo',
                'IP_Item_30_Tiempo'
            ]
            print(f"      üîß Definiendo manualmente: {Items_Conservadores_Tiempo}")

print("\n" + "="*60)

for Nombre_df, df in dfs_Finales.items():
    print(f"\nüìä DATAFRAME: {Nombre_df}")
    print("-" * 50)
    
    # 2. Verificar existencia de columnas de √≠tems conservadores.
    print("2. VERIFICACI√ìN DE COLUMNAS CONSERVADORAS:")
    
    if 'Items_Conservadores' in globals():
        Columnas_Encontradas = []
        Columnas_Faltantes = []
        
        for Item in Items_Conservadores:
            if Item in df.columns:
                Columnas_Encontradas.append(Item)
            else:
                Columnas_Faltantes.append(Item)
        
        print(f"   ‚úÖ Columnas encontradas: {len(Columnas_Encontradas)}")
        print(f"   ‚ùå Columnas faltantes: {len(Columnas_Faltantes)}")
        
        if Columnas_Faltantes:
            print(f"      Faltantes: {Columnas_Faltantes}")
    
    # 3. Verificar la inversi√≥n de escala.
    print("\n3. VERIFICACI√ìN DE INVERSI√ìN DE ESCALA:")
    
    if 'Items_Conservadores' in globals() and len(Columnas_Encontradas) > 0:
        # Tomar una muestra de la primera columna conservadora.
        Primera_Columna = Columnas_Encontradas[0]
        
        # Verificar rango de valores despu√©s de la inversi√≥n.
        Valores_Unicos = df[Primera_Columna].dropna().unique()
        Valores_Unicos_Ordenados = sorted([v for v in Valores_Unicos if pd.notna(v)])
        
        print(f"   Muestra en {Primera_Columna}:")
        print(f"   Valores √∫nicos despu√©s de inversi√≥n: {Valores_Unicos_Ordenados}")
        
        # Verificar si los valores est√°n en el rango esperado [1-5].
        Valores_Fuera_Rango = [v for v in Valores_Unicos_Ordenados if v < 1 or v > 5]
        if Valores_Fuera_Rango:
            print(f"   ‚ö†Ô∏è  Valores fuera de rango [1-5]: {Valores_Fuera_Rango}")
        else:
            print(f"   ‚úÖ Todos los valores est√°n en rango [1-5]")
        
        # Verificar tipos de datos.
        Tipo_Datos = str(df[Primera_Columna].dtype)
        print(f"   Tipo de datos: {Tipo_Datos}")
        
        if 'float' in Tipo_Datos:
            print("   ‚úÖ Conversi√≥n a float exitosa")
        else:
            print("   ‚ö†Ô∏è  No es tipo float")
    
    # 4. Verificar c√°lculo del √çndice de Conservadurismo.
    print("\n4. VERIFICACI√ìN DEL √çNDICE DE CONSERVADURISMO:")
    
    if 'Indice_Conservadurismo' in df.columns:
        print("   ‚úÖ Columna 'Indice_Conservadurismo' creada")
        
        # Estad√≠sticas descriptivas.
        Estadisticas_Indice = df['Indice_Conservadurismo'].describe()
        print(f"   Estad√≠sticas descriptivas:")
        print(f"     Count: {Estadisticas_Indice['count']:.0f}")
        print(f"     Mean: {Estadisticas_Indice['mean']:.3f}")
        print(f"     Std: {Estadisticas_Indice['std']:.3f}")
        print(f"     Min: {Estadisticas_Indice['min']:.3f}")
        print(f"     Max: {Estadisticas_Indice['max']:.3f}")
        
        # Verificar rango esperado.
        if Estadisticas_Indice['min'] >= 1 and Estadisticas_Indice['max'] <= 5:
            print("   ‚úÖ √çndice en rango esperado [1-5]")
        else:
            print(f"   ‚ö†Ô∏è  √çndice fuera de rango esperado: [{Estadisticas_Indice['min']:.3f}, {Estadisticas_Indice['max']:.3f}]")
        
        # Verificar valores NaN.
        Valores_NaN = df['Indice_Conservadurismo'].isna().sum()
        Total_Filas = len(df)
        Porcentaje_NaN = (Valores_NaN / Total_Filas) * 100
        print(f"   Valores NaN: {Valores_NaN} ({Porcentaje_NaN:.1f}%)")
        
        # Muestra de valores del √≠ndice.
        Muestra_Indices = df['Indice_Conservadurismo'].dropna().head(10).tolist()
        print(f"   Muestra de valores: {[round(v, 3) for v in Muestra_Indices]}")
        
    else:
        print("   ‚ùå Columna 'Indice_Conservadurismo' NO encontrada")
    
    # 5. Verificar √çndice de Tiempo de Conservadurismo.
    print("\n5. VERIFICACI√ìN DEL √çNDICE DE TIEMPO:")
    
    if 'Indice_Conservadurismo_Tiempo' in df.columns:
        print("   ‚úÖ Columna 'Indice_Conservadurismo_Tiempo' creada")
        
        # Estad√≠sticas descriptivas.
        Estadisticas_Tiempo = df['Indice_Conservadurismo_Tiempo'].describe()
        print(f"   Estad√≠sticas descriptivas:")
        print(f"     Count: {Estadisticas_Tiempo['count']:.0f}")
        print(f"     Mean: {Estadisticas_Tiempo['mean']:.3f}")
        print(f"     Std: {Estadisticas_Tiempo['std']:.3f}")
        print(f"     Min: {Estadisticas_Tiempo['min']:.3f}")
        print(f"     Max: {Estadisticas_Tiempo['max']:.3f}")
        
        # Verificar valores negativos (no deber√≠an existir).
        Valores_Negativos = (df['Indice_Conservadurismo_Tiempo'] < 0).sum()
        if Valores_Negativos > 0:
            print(f"   ‚ö†Ô∏è  Valores negativos: {Valores_Negativos}")
        else:
            print("   ‚úÖ No hay valores negativos")
        
        # Valores NaN.
        Valores_NaN_Tiempo = df['Indice_Conservadurismo_Tiempo'].isna().sum()
        Porcentaje_NaN_Tiempo = (Valores_NaN_Tiempo / Total_Filas) * 100
        print(f"   Valores NaN: {Valores_NaN_Tiempo} ({Porcentaje_NaN_Tiempo:.1f}%)")
        
    else:
        print("   ‚ùå Columna 'Indice_Conservadurismo_Tiempo' NO encontrada")
    
    # 6. Verificaci√≥n manual de c√°lculo de √≠ndice.
    print("\n6. VERIFICACI√ìN MANUAL DEL C√ÅLCULO:")
    
    if ('Indice_Conservadurismo' in df.columns and 
        'Items_Conservadores' in globals() and 
        len(Columnas_Encontradas) > 0):
        
        # Tomar una fila aleatoria para verificaci√≥n manual.
        Fila_Muestra = df.dropna(subset=['Indice_Conservadurismo']).iloc[0]
        ID_Muestra = Fila_Muestra['ID']
        Indice_Calculado = Fila_Muestra['Indice_Conservadurismo']
        
        print(f"   Verificando ID: {ID_Muestra}")
        print(f"   √çndice calculado: {Indice_Calculado:.4f}")
        
        # Calcular manualmente.
        Valores_Items = []
        for Item in Columnas_Encontradas:
            Valor = Fila_Muestra[Item]
            if pd.notna(Valor):
                Valores_Items.append(Valor)
                print(f"     {Item}: {Valor}")
        
        if len(Valores_Items) > 0:
            Indice_Manual = sum(Valores_Items) / len(Valores_Items)
            print(f"   √çndice manual: {Indice_Manual:.4f}")
            
            Diferencia = abs(Indice_Calculado - Indice_Manual)
            if Diferencia < 0.001:
                print("   ‚úÖ C√°lculo correcto")
            else:
                print(f"   ‚ö†Ô∏è  Diferencia: {Diferencia:.4f}")
    
    # 7. Distribuci√≥n por categor√≠as.
    print("\n7. DISTRIBUCI√ìN POR CATEGOR√çAS PASO:")
    
    if ('Indice_Conservadurismo' in df.columns and 
        'Categoria_PASO_2023' in df.columns):
        
        Distribucion_Categorias = df.groupby('Categoria_PASO_2023')['Indice_Conservadurismo'].agg(['count', 'mean', 'std'])
        
        print("   Categor√≠a               | Count | Mean  | Std   |")
        print("   " + "-"*50)
        
        for Categoria, Stats in Distribucion_Categorias.iterrows():
            if pd.notna(Stats['mean']):
                print(f"   {Categoria:<22} | {Stats['count']:5.0f} | {Stats['mean']:5.2f} | {Stats['std']:5.2f} |")
    
    # 8. COMPARACI√ìN CON √çNDICE DE PROGRESISMO.
    print("\n8. COMPARACI√ìN PROGRESISMO VS CONSERVADURISMO:")
    
    if ('Indice_Conservadurismo' in df.columns and 
        'Indice_Progresismo' in df.columns):
        
        # Calcular correlaci√≥n.
        Correlacion = df['Indice_Conservadurismo'].corr(df['Indice_Progresismo'])
        print(f"   Correlaci√≥n entre √≠ndices: {Correlacion:.3f}")
        
        if Correlacion < -0.5:
            print("   ‚úÖ Correlaci√≥n negativa fuerte (esperado)")
        elif Correlacion < 0:
            print("   ‚ö†Ô∏è  Correlaci√≥n negativa d√©bil")
        else:
            print("   ‚ùå Correlaci√≥n positiva (problema)")
        
        # Comparar medias por categor√≠a.
        print("\n   Comparaci√≥n de medias por categor√≠a:")
        print("   Categor√≠a               | Progres. | Conserv. | Diferencia |")
        print("   " + "-"*65)
        
        Comparacion = df.groupby('Categoria_PASO_2023')[['Indice_Progresismo', 'Indice_Conservadurismo']].mean()
        
        for Categoria, Stats in Comparacion.iterrows():
            if pd.notna(Stats['Indice_Progresismo']) and pd.notna(Stats['Indice_Conservadurismo']):
                Diferencia = Stats['Indice_Progresismo'] - Stats['Indice_Conservadurismo']
                print(f"   {Categoria:<22} | {Stats['Indice_Progresismo']:8.2f} | {Stats['Indice_Conservadurismo']:8.2f} | {Diferencia:10.2f} |")

print("\n" + "="*80)
print("‚úÖ VERIFICACI√ìN DEL √çNDICE DE CONSERVADURISMO COMPLETADA")
print("="*80)

# 9. Verificaci√≥n adicional: Coherencia ideol√≥gica.
print("\n" + "="*80)
print("VERIFICACI√ìN ADICIONAL: COHERENCIA IDEOL√ìGICA")
print("="*80)

for Nombre_df, df in dfs_Finales.items():
    print(f"\nüîç DATAFRAME: {Nombre_df}")
    print("-" * 40)
    
    if ('Indice_Conservadurismo' in df.columns and 
        'Indice_Progresismo' in df.columns and
        'Categoria_PASO_2023' in df.columns):
        
        print("Verificando coherencia ideol√≥gica por categor√≠a:")
        print("(Las categor√≠as de izquierda deber√≠an tener mayor progresismo)")
        print("(Las categor√≠as de derecha deber√≠an tener mayor conservadurismo)")
        
        # Categor√≠as esperadas de izquierda a derecha.
        Orden_Ideologico = ['Left_Wing', 'Progressivism', 'Centre', 
                           'Moderate_Right_A', 'Moderate_Right_B', 
                           'Right_Wing_Libertarian']
        
        print("\n   √çndices promedio por categor√≠a (izquierda ‚Üí derecha):")
        for Categoria in Orden_Ideologico:
            if Categoria in df['Categoria_PASO_2023'].values:
                df_Cat = df[df['Categoria_PASO_2023'] == Categoria]
                Prog_Medio = df_Cat['Indice_Progresismo'].mean()
                Cons_Medio = df_Cat['Indice_Conservadurismo'].mean()
                
                if pd.notna(Prog_Medio) and pd.notna(Cons_Medio):
                    Tendencia = "Progresista" if Prog_Medio > Cons_Medio else "Conservadora"
                    print(f"   {Categoria:<22}: Prog={Prog_Medio:.2f}, Cons={Cons_Medio:.2f} ‚Üí {Tendencia}")
        
        # Verificar tendencia general.
        print("\n   Verificaci√≥n de tendencia esperada:")
        
        # Izquierda deber√≠a ser m√°s progresista.
        Categorias_Izquierda = ['Left wing', 'Progressivism']
        Categorias_Derecha = ['Moderate right A', 'Moderate right B', 'Right wing libertarian']
        
        Prog_Izquierda = df[df['Categoria_PASO_2023'].isin(Categorias_Izquierda)]['Indice_Progresismo'].mean()
        Prog_Derecha = df[df['Categoria_PASO_2023'].isin(Categorias_Derecha)]['Indice_Progresismo'].mean()
        
        Cons_Izquierda = df[df['Categoria_PASO_2023'].isin(Categorias_Izquierda)]['Indice_Conservadurismo'].mean()
        Cons_Derecha = df[df['Categoria_PASO_2023'].isin(Categorias_Derecha)]['Indice_Conservadurismo'].mean()
        
        print(f"   Izquierda - Progresismo: {Prog_Izquierda:.2f}")
        print(f"   Derecha - Progresismo: {Prog_Derecha:.2f}")
        print(f"   Izquierda - Conservadurismo: {Cons_Izquierda:.2f}")
        print(f"   Derecha - Conservadurismo: {Cons_Derecha:.2f}")
        
        if (pd.notna(Prog_Izquierda) and pd.notna(Prog_Derecha) and 
            pd.notna(Cons_Izquierda) and pd.notna(Cons_Derecha)):
            
            if Prog_Izquierda > Prog_Derecha:
                print("   ‚úÖ Izquierda m√°s progresista que derecha")
            else:
                print("   ‚ö†Ô∏è  Izquierda NO m√°s progresista que derecha")
            
            if Cons_Derecha > Cons_Izquierda:
                print("   ‚úÖ Derecha m√°s conservadora que izquierda")
            else:
                print("   ‚ö†Ô∏è  Derecha NO m√°s conservadora que izquierda")

print("\n" + "="*80)
print("‚úÖ VERIFICACI√ìN DE COHERENCIA IDEOL√ìGICA COMPLETADA")
print("="*80)
```
