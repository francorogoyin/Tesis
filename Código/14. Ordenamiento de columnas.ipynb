```python
print("="*80)
print("VERIFICACIÓN: ORDENAMIENTO DE COLUMNAS")
print("="*80)

# 1. Verificar errores de sintaxis en la lista de orden.
print("1. VERIFICACIÓN DE ERRORES DE SINTAXIS:")

# Detectar errores comunes en la lista Orden_Columnas.
if 'Orden_Columnas' in globals():
    Lista_Orden = globals()['Orden_Columnas']
    print(f"   ✅ Lista 'Orden_Columnas' encontrada con {len(Lista_Orden)} elementos")
    
    # Buscar errores de sintaxis comunes.
    Errores_Encontrados = []
    
    for i, Columna in enumerate(Lista_Orden):
        # Error 1: Líneas sin coma.
        if i < len(Lista_Orden) - 1:  # No es el último elemento.
            Linea_Actual = str(Columna)
            if "'" in Linea_Actual and not Linea_Actual.strip().endswith(("'", '"')):
                Errores_Encontrados.append(f"Línea {i+1}: Posible error de sintaxis en '{Columna}'")
    
    # Buscar duplicados.
    Columnas_Unicas = set()
    Duplicados = []
    for Columna in Lista_Orden:
        if Columna in Columnas_Unicas:
            Duplicados.append(Columna)
        else:
            Columnas_Unicas.add(Columna)
    
    if Errores_Encontrados:
        print("   ⚠️  Errores de sintaxis detectados:")
        for Error in Errores_Encontrados[:5]:  # Mostrar máximo 5.
            print(f"     {Error}")
    else:
        print("   ✅ No se detectaron errores de sintaxis obvios")
    
    if Duplicados:
        print(f"   ⚠️  Columnas duplicadas: {Duplicados}")
    else:
        print("   ✅ No hay columnas duplicadas en la lista")

else:
    print("   ❌ Lista 'Orden_Columnas' no encontrada")
    # Crear una lista básica para continuar la verificación.
    Orden_Columnas = ['ID', 'Edad', 'Genero', 'Region']
    print("   🔧 Usando lista básica para verificación")

# 2. Verificar columnas de cambio de opinión.
print("\n2. VERIFICACIÓN DE COLUMNAS DE CAMBIO DE OPINIÓN:")

if 'Columnas_Cambio_Opinion' in globals():
    Lista_Cambio = globals()['Columnas_Cambio_Opinion']
    print(f"   ✅ Lista definida con {len(Lista_Cambio)} columnas:")
    for Columna in Lista_Cambio:
        print(f"     - {Columna}")
else:
    print("   ❌ Lista 'Columnas_Cambio_Opinion' no encontrada")
    Lista_Cambio = ['Cambio_Opinion_Pro', 'Cambio_Tiempo_Pro']
    print("   🔧 Usando lista reducida para verificación")

print("\n" + "="*60)

for Nombre_df, df in dfs_Finales.items():
    print(f"\n📋 DATAFRAME: {Nombre_df}")
    print("-" * 50)
    
    Columnas_Antes = len(df.columns)
    
    # 3. Verificar adición de columnas de cambio de opinión.
    print("3. VERIFICACIÓN DE COLUMNAS AÑADIDAS:")
    
    Columnas_Cambio_Encontradas = []
    Columnas_Cambio_Faltantes = []
    
    for Columna in Lista_Cambio:
        if Columna in df.columns:
            Columnas_Cambio_Encontradas.append(Columna)
            # Verificar que sean todas 0.
            Valores_Unicos = df[Columna].unique()
            if len(Valores_Unicos) == 1 and Valores_Unicos[0] == 0:
                print(f"   ✅ {Columna}: Añadida correctamente (todos 0)")
            else:
                print(f"   ⚠️  {Columna}: Valores no son todos 0: {Valores_Unicos}")
        else:
            Columnas_Cambio_Faltantes.append(Columna)
    
    if Columnas_Cambio_Faltantes:
        print(f"   ❌ Columnas faltantes: {Columnas_Cambio_Faltantes}")
    
    # 4. Verificar orden de columnas.
    print("\n4. VERIFICACIÓN DE ORDEN DE COLUMNAS:")
    
    # Verificar que las primeras columnas estén en el orden correcto.
    Primeras_10_Esperadas = [col for col in Orden_Columnas[:10] if col in df.columns]
    Primeras_10_Reales = list(df.columns[:len(Primeras_10_Esperadas)])
    
    if Primeras_10_Esperadas == Primeras_10_Reales:
        print("   ✅ Primeras 10 columnas en orden correcto")
    else:
        print("   ⚠️  Orden de primeras columnas no coincide:")
        print(f"     Esperado: {Primeras_10_Esperadas}")
        print(f"     Real:     {Primeras_10_Reales}")
    
    # Verificar que ID esté primero.
    if len(df.columns) > 0 and df.columns[0] == 'ID':
        print("   ✅ 'ID' está en la primera posición")
    else:
        print(f"   ⚠️  Primera columna es '{df.columns[0]}' en lugar de 'ID'")
    
    # 5. Verificar agrupaciones temáticas.
    print("\n5. VERIFICACIÓN DE AGRUPACIONES TEMÁTICAS:")
    
    # Definir grupos temáticos esperados.
    Grupos_Tematicos = {
        'Demográficos': ['Edad', 'Genero', 'Estrato_Social', 'Nivel_Educativo'],
        'Geográficos': ['Provincia', 'Region'],
        'Electorales': ['Voto_2019', 'Voto_PASO_2023', 'Candidato_PASO_2023'],
        'Redes_Sociales': ['Red_Social_Twitter', 'Red_Social_Facebook', 'Red_Social_Instagram'],
        'Medios_Prensa': ['Medios_Prensa_Clarin', 'Medios_Prensa_La_Nacion', 'Medios_Prensa_Infobae']
    }
    
    for Grupo, Columnas_Grupo in Grupos_Tematicos.items():
        Columnas_Encontradas = [col for col in Columnas_Grupo if col in df.columns]
        if Columnas_Encontradas:
            # Verificar si están agrupadas (posiciones cercanas).
            Posiciones = [df.columns.get_loc(col) for col in Columnas_Encontradas]
            Posiciones.sort()
            
            # Calcular dispersión.
            if len(Posiciones) > 1:
                Dispersión = max(Posiciones) - min(Posiciones)
                Esperada = len(Posiciones) - 1  # Posiciones consecutivas.
                
                if Dispersión <= Esperada * 3:  # Tolerancia 3x.
                    print(f"   ✅ {Grupo}: Columnas agrupadas (dispersión {Dispersión})")
                else:
                    print(f"   ⚠️  {Grupo}: Columnas dispersas (dispersión {Dispersión})")
            else:
                print(f"   ✅ {Grupo}: {len(Columnas_Encontradas)} columna encontrada")
        else:
            print(f"   ❌ {Grupo}: No se encontraron columnas")
    
    # 6. Verificar nomenclatura de redes sociales.
    print("\n6. VERIFICACIÓN DE NOMENCLATURA - REDES SOCIALES:")
    
    Redes_Esperadas = ['Twitter', 'Facebook', 'Instagram', 'Threads', 'Tiktok', 
                      'Youtube', 'Whatsapp', 'Telegram']
    
    # Verificar nombres antiguos (no deberían existir).
    Nombres_Antiguos = [red for red in Redes_Esperadas if red in df.columns]
    Nombres_Nuevos = [f"Red_Social_{red}" for red in Redes_Esperadas if f"Red_Social_{red}" in df.columns]
    
    print(f"   Nombres antiguos encontrados: {len(Nombres_Antiguos)}")
    if Nombres_Antiguos:
        print(f"     ⚠️  {Nombres_Antiguos}")
    
    print(f"   Nombres nuevos encontrados: {len(Nombres_Nuevos)}")
    if len(Nombres_Nuevos) > 0:
        print(f"     ✅ {Nombres_Nuevos[:5]}...")  # Mostrar solo primeros 5.
    
    # 7. Verificar nomenclatura de medios de prensa.
    print("\n7. VERIFICACIÓN DE NOMENCLATURA - MEDIOS DE PRENSA:")
    
    Medios_Esperados = ['Clarin', 'La_Nacion', 'Infobae', 'Pagina_12', 'Perfil']
    
    # Verificar nombres antiguos.
    Medios_Antiguos = [medio for medio in Medios_Esperados if medio in df.columns]
    Medios_Nuevos = [f"Medios_Prensa_{medio}" for medio in Medios_Esperados 
                    if f"Medios_Prensa_{medio}" in df.columns]
    
    print(f"   Nombres antiguos encontrados: {len(Medios_Antiguos)}")
    if Medios_Antiguos:
        print(f"     ⚠️  {Medios_Antiguos}")
    
    print(f"   Nombres nuevos encontrados: {len(Medios_Nuevos)}")
    if len(Medios_Nuevos) > 0:
        print(f"     ✅ {Medios_Nuevos[:5]}...")
    
    # 8. VERIFICACIÓN MANUAL DE CASOS ESPECÍFICOS.
    print("\n8. VERIFICACIÓN MANUAL - ESTRUCTURA DE COLUMNAS:")
    
    # Mostrar primeras 15 columnas.
    print("   Primeras 15 columnas:")
    for i, col in enumerate(df.columns[:15], 1):
        print(f"     {i:2}. {col}")
    
    # Mostrar últimas 10 columnas.
    print(f"\n   Últimas 10 columnas (de {len(df.columns)} totales):")
    for i, col in enumerate(df.columns[-10:], len(df.columns)-9):
        print(f"     {i:2}. {col}")
    
    # 9. Verificar integridad (columnas perdidas).
    print("\n9. VERIFICACIÓN DE INTEGRIDAD:")
    
    Columnas_Despues = len(df.columns)
    
    print(f"   Columnas antes del ordenamiento: ~{Columnas_Antes}")
    print(f"   Columnas después del ordenamiento: {Columnas_Despues}")
    
    # Verificar si hay pérdida significativa de columnas.
    if Columnas_Despues < Columnas_Antes * 0.95:  # Pérdida > 5%.
        print(f"   ⚠️  Posible pérdida de columnas ({Columnas_Antes - Columnas_Despues})")
    else:
        print("   ✅ No hay pérdida significativa de columnas")
    
    # 10. Verificar columnas críticas.
    print("\n10. VERIFICACIÓN DE COLUMNAS CRÍTICAS:")
    
    Columnas_Criticas = ['ID', 'Edad', 'Genero', 'Region', 'Categoria_PASO_2023']
    
    for Columna in Columnas_Criticas:
        if Columna in df.columns:
            Posicion = df.columns.get_loc(Columna) + 1
            print(f"   ✅ {Columna}: Posición {Posicion}")
        else:
            print(f"   ❌ {Columna}: No encontrada")
    
    # 11. Análisis de distribución de tipos de columnas.
    print("\n11. ANÁLISIS DE TIPOS DE COLUMNAS:")
    
    # Contar tipos de columnas por prefijo.
    Contadores = {
        'Variables demográficas': 0,
        'Variables dummy': 0,
        'Redes sociales': 0,
        'Medios de prensa': 0,
        'Items IP': 0,
        'Variables calculadas': 0,
        'Otras': 0
    }
    
    for Columna in df.columns:
        if Columna in ['Edad', 'Genero', 'Estrato_Social', 'Nivel_Educativo', 'Region']:
            Contadores['Variables demográficas'] += 1
        elif '_' in Columna and any(Columna.startswith(f"{var}_") for var in 
                                   ['Genero', 'Region', 'Nivel_Educativo', 'Estrato_Social']):
            Contadores['Variables dummy'] += 1
        elif Columna.startswith('Red_Social_'):
            Contadores['Redes sociales'] += 1
        elif Columna.startswith('Medios_Prensa_'):
            Contadores['Medios de prensa'] += 1
        elif Columna.startswith('IP_Item_'):
            Contadores['Items IP'] += 1
        elif Columna.startswith(('Indice_', 'Tiempos_', 'Cambio_')):
            Contadores['Variables calculadas'] += 1
        else:
            Contadores['Otras'] += 1
    
    for Tipo, Cantidad in Contadores.items():
        Porcentaje = (Cantidad / len(df.columns)) * 100
        print(f"   {Tipo:<20}: {Cantidad:3} ({Porcentaje:5.1f}%)")

print("\n" + "="*80)
print("✅ VERIFICACIÓN DE ORDENAMIENTO COMPLETADA")
print("="*80)

# 12. Resumen ejecutivo.
print("\n" + "="*80)
print("RESUMEN EJECUTIVO - ORDENAMIENTO DE COLUMNAS")
print("="*80)

for Nombre_df, df in dfs_Finales.items():
    print(f"\n📊 DATAFRAME: {Nombre_df}")
    
    # Verificaciones clave.
    ID_Primera = df.columns[0] == 'ID' if len(df.columns) > 0 else False
    
    # Contar prefijos aplicados.
    Redes_Renombradas = sum(1 for col in df.columns if col.startswith('Red_Social_'))
    Medios_Renombrados = sum(1 for col in df.columns if col.startswith('Medios_Prensa_'))
    
    # Columnas críticas presentes.
    Criticas_Presentes = sum(1 for col in ['ID', 'Edad', 'Genero', 'Region', 'Categoria_PASO_2023'] 
                           if col in df.columns)
    
    print(f"   📋 Total de columnas: {len(df.columns)}")
    print(f"   🔢 ID en primera posición: {'✅' if ID_Primera else '❌'}")
    print(f"   📱 Redes sociales renombradas: {Redes_Renombradas}")
    print(f"   📰 Medios de prensa renombrados: {Medios_Renombrados}")
    print(f"   🎯 Columnas críticas presentes: {Criticas_Presentes}/5")
    
    # Estado general.
    if ID_Primera and Criticas_Presentes >= 4 and (Redes_Renombradas > 0 or Medios_Renombrados > 0):
        print(f"   ✅ Dataset correctamente estructurado para análisis")
    else:
        print(f"   ⚠️  Dataset requiere ajustes menores")

print("\n" + "="*80)
print("✅ ANÁLISIS DE ESTRUCTURA FINAL COMPLETADO")
print("="*80)
```
